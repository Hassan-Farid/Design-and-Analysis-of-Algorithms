# Merge Sort:
MergeSort is a sorting method that uses the divide and conquer approach to sort the elements in an array. For Merge Sort we have:

### Divide Step:
The divide step in merge sort breaks the given array of size n into two halves i.e. n/2 each

### Conquer Step:
THe conquer step recursively divides the sub-problems into further sub-problems until individual elements are left.

### Combine Step:
The combine step merges the individual items left by conquer step in the right order to create a sorted array.

## Psuedocode:
The psuedocode has two parts i.e. one main algorithm that runs the main merge sort algorithm and the other one a sub-algorithm to merge the values for the combine step in the algorithm.

### Merge Sort Algorithm Psuedocode:
<div style="background-color:rgba(0, 0, 0, 0.0470588); padding:10px 0;font-family:monospace;">
    <font>MergeSort(A,p,r):</font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; <font> if p $<$ r: </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> do q $\leftarrow$ $\lfloor$(p + r)/2 $\rfloor$ </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> do i $\leftarrow$ j - 1 </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> MergeSort(A,p,q) </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> MergeSort(A,q+1,r) </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> Merge(A,p,q,r) </font><br>
</div>

#### Explanation:
MergeSort algorithm as described above uses the divide and conquer approach to get its result. Here the parameters p and r provided indicate the start and ending index of the array/sub-array on which the merge sort procedure is to be applied. Now once these inputs are provided and the merge sort procedure is called, the algorithm first checks whether p < r because if its not then that means the start index of the array is greater than the ending index, which in turn means that the array has ended. Now, once the condition is satisfied, we compute the middle index of the array so as to divide the array into two equal parts. Once middle element is found, we apply merge sort procedures on sub-arrays individually recursively as indicated to break each sub-array into more sub-array and continue the divide and conquer step. Once all divide and conquer steps are applied recursively, we call the Merge sub-algorithm, which merges all the divided items in the right order and creates a sorted array.

### Merge Sub-Algorithm Psuedocode:
<div style="background-color:rgba(0, 0, 0, 0.0470588); padding:10px 0;font-family:monospace;">
    <font>Merge(A,p,q,r):</font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; <font> do n$_1$ $\leftarrow$ q - p + 1 </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; <font> do n$_2$ $\leftarrow$ r - q </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; <font> Let L[1 ... n$_1$ + 1] and R[1 ... n$_2$ + 1] </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; <font> for i from 1 to n$_1$: </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp;<font> do L[i] $\leftarrow$ A[p+i-1] </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; <font> for j from 1 to n$_2$: </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> do R[j] $\leftarrow$ A[q+j] </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; <font> L[n$_1$ + 1] $\leftarrow$ R[n$_2$ + 1] $\leftarrow$ $\infty$ </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; <font> i $\leftarrow$ j $\leftarrow$ 1 </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; <font> for k from p to r: </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> if L[i] $\leq$ R[j]:</font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> do A[k] $\leftarrow$ L[i] </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> do i $\leftarrow$ i+1 </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> else: </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> do A[k] $\leftarrow$ R[j] </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> do j $\leftarrow$ j+1 </font><br>
</div>

#### Explanation:
The Merge procedure takes four parameters i.e. the array to be sorted, its starting index, its middle index and its last index respectively. Once it knows the values of these parameters, it creates two subarrays i.e. L[p ... q] and R[q+1 ... r] indicating the left and right parts of the array to be merged into. once that is assigned, we assign sentinal value $\infty$ and assign it to the last index of both of these created arrays so that the loop for assigning values terminates once both arrays have filled up half elements of the array to be merged into. Once the sub-arrays have been allocated, the Mergesort procedure. Once this is done, we start the loop to merge the values in the right order, which merges based on the following to conditions:

* If L[i] $\leq$ R[j] i.e the current value in L is less than equal to the current value in R, put the value of L in the merged array.
* Otherwise if R[j] < L[i] i.e. the current value of L is greater than the current value in R, put the value of R in the merged array.

At some point in the merging, the condition will arise where there will be no elements one of the arrays (suppose L), then the sentinal value comes into play. Since no value can be greater than $\infty$, thus for each iteration, the value in R will be merged until the loop is complete. the result we get from the merging procedure is a sorted array.

#### Loop Invariant:
As we can see we use a loop in the merge procedure to merge all the items of a list thus we can prove a loop invariant on this loop so as to show that the result obtained from the merge procedure is a sorted array, we have:

##### Initialization:
Before the start of the loop, the merged array is empty. Since the array is empty it is already sorted hypothetically.

##### Maintainence:
At start of each iteration, the subarray A[p ... k-1] contains k-p smallest elements of L[1 ... n$_1$ + 1] and R[1 ... n$_2$ + 1] in sorted order. Moreover, L[i] and R[j] are the smallest elements of their arrays that have not been copied back into A.

##### Termination:
We know that every step in the loop 1 item is stored in the merged array A in sorted order. Since the loop runs (r - p + 1) times, thus the total number of items stored in sorted order in the array A after all iterations will be (r - p + 1), which were the required number of items to be sorted. Hence, the termination of the loop would provide us with a merged sorted array A. The termination procedure is controlled using a sentinel value ($\infty$ for case of ascending order and -$\infty$ for descending order)

## Recurrence for Merge Sort:
Since Merge Sort uses recursive functions, thus, it has a recurrence relation, which is given below:

> T(n) = O(1) if n = 1

> T(n) = 2T(n/2) + O(n) 

where,

* T(n) = Running Time for input of size n
* 2T(n/2) = Running Time for the two sub-arrays of size n/2 created from the divide step
* O(n) = Linear Time taken to perform extra work (dividing and merging the individual items)

### Solving the Recurrence Relation:
Since the recurrence relation matches the format for the Master's Theorem i.e. T(n) = aT(n/b) + f(n), thus, on comparing we have:

* a = 2
* b = 2
* f(n) = n

Now, for the value of n$^{log_b a}$:

> n$^{log_b a}$ = n$^{log_2 2}$ = n$^1$ = n = f(n)

Since, we have f(n) = $\theta$(n$^{log_b a}$), thus, we apply case II of the Master's theorem with 
k = 0:

> T(n) = O(n$^{log_b a}$ * log$_2^{k + 1}$ n) = O(n * log$_2^1$ n) = O(nlog n)

which is the required running time for our algorithm

### Space complexity for Merge Sort:
The Merge Sort is an outplace algorithm which divides the arrays into sub arrays in place but merges them outplace by creating arrays of atmost size n (where n is the size of input array: r - p + 1). Thus, the space complexity of merge sort (both overall and auxilary) is:

> S(n) = O(n)
