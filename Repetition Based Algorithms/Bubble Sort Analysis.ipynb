# Problem: Sorting

### Input:
A sequence of n - elements which is to be sorted

### Output:
A permutation of the sequence where each element on the right is less than the left one (sorted in ascending order)

### Method:
We are using the Bubble sort method to sort the elements of the sequence.

### Pseudocode:
<div style="background-color:rgba(0, 0, 0, 0.0470588); padding:10px 0;font-family:monospace;">
    <font>BubbleSort(A):</font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; <font> for i $\leftarrow$ 1 to A.length - 1: </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> for j $\leftarrow$ A.length downto i+1:  </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> if A[j] $<$  A[j - 1]: </font><br>
    &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; <font> exchange A[j] with A[j -1] </font><br>
</div>

### Explanation:
Bubblesort is an inefficient algorithm which works by repeatedly swapping adjacent elements that are out of order. We use nested looping to compare each element and swap them in the right sorted order.

### Loop Invariant:
For loop invariant of bubble sort, we have:

#### Initialization:
At the start of the loop, there are no elements in the array, thus it is already sorted. Now when the inner loop starts, we compare the 1$^{st}$ element in A with the rest of the elements and get the smallest value in A as A[1].

#### Maintainence:
For the j$^{th}$ element of the loop, there are A[1 ... j-1] already in the sorted order. Now when inner loop starts, we compare j$^{th}$ element in A with the rest of elements and get smallest value in A as A[j].

#### Termination:
By the end of the loop, j pointer has reached the i+1 value which is compared to the i$^{th}$ value, ultimately sorting the whole array A (each time the value of i increments and terminates with reaching A.length). 

### Running Time of Bubble Sort:
For the running time of Bubble sort, we have the following observation:

* __For i = 1:__ j = [n, n-1, ... 2, 1] => (n - 1) comparisons
* __For i = 2:__ j = [n, n-1, ... 3, 2] => (n - 2) comparisons
* __For i = 3:__ j = [n, n-1, ... 4, 3] => (n - 3) comparisons
.
.
.
.
* __For i = n-1:__ j = [n, n-1] => 1 comparisons
* __For i = n:__ j = [n] => 0 comparisons

Summing up all the number of comparisons, we get:

> T(n) = (n - 1) + (n - 2) + ... + 2 + 1 = n(n - 1)/2 = (n$^2$ - n)/2 (Using Arithmetic Series Formula)

One thing to note here is that it doesn't matter whether the array is sorted already or not, the nested loop always runs with O(1) comparisons, thus, both best and worst case running times for bubble sort are:

> T(n) = O(n$^2$)

which is the required running time

### Space Complexity for Bubble Sort:
Bubble Sort is an inplace algorithm, meaning that the array is sorted without occupying more space than that of the allocated n-sized array. Thus, we have for overall space complexity of bubble sort:

> S(n) = O(n)

On the other hand, since it is inplace, its auxilary space complexity is:

> S(n) = O(1)
