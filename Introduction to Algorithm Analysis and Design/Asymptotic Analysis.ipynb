# Asymptotic Analysis:
When performing analysis on a certain algorithm, there are many factors that change the performance of the algorithm on a machine. These may include the hardware, compiler, programming language,etc. being used. Computer Scientists observed that this dependency on the system's componenets is pretty difficult to analyze or observe, hence the process of "Asymptotic Analysis" was invented i.e. analyzing algorithms based on the input size 'n' as n $\rightarrow \infty$. In asymptotic analysis, we compress the running time or space occupied expression to a single degree of n, which indicates the upper/lower bound for the particular algorithm. The degree of n is taken on the basis of the asymptotic notation being used.

# Asymptotic Notations:
There are three common asymptotic notations used for performing asymptotic analysis of algorithms:

## Big O-Notation:
A function f(n) is said to be O(g(n)) if and only if there exists positive constant c such that:

> 0 $\leq$ f(n) $\leq$ cg(n) ; $\forall$ n $\epsilon$ N, n $\geq$ n$_o$

We illustrate a simple graph for this expression using a simple example below: 

```python
#Importing libraries
from matplotlib import pyplot as plt

#Consider the following two functions representing the running times for some functions
def f(n):
    return n**2

def g(n):
    return n**3

#Creating list of values yielded based on the functions
x_vals = list(range(50))
f = [f(x) for x in x_vals]
g = [g(x) for x in x_vals]

#Plotting functions using matplotlib
plt.plot(x_vals, g, color='r', label='g(n)')
plt.plot(x_vals, f, color='b', label='f(n)')

plt.xlabel('Input size (n)')
plt.ylabel('Running Time T(n)')
plt.legend()

plt.show()
```

### Example:
In the above example, we took two functions with running times n$^2$ and n$^3$ respectively. Now, as the size of input increase, it is clear that the output for n$^3$ will increase more rapidly than the output of n$^2$. Since, we have g(n) as an upper bound over f(n), thus, we can say for the provided example:

> f(n) = O(g(n)) $\forall$ n > 1

## Big $\Omega$-Notation:
A function f(n) is said to be $\Omega$(g(n)) if and only if there exists positive constant c such that:

> 0 $\leq$ cg(n) $\leq$ f(n) ; $\forall$ n $\epsilon$ N, n $\geq$ n$_o$

We illustrate a simple graph for this expression using a simple example below: 

```python
#Importing libraries
from matplotlib import pyplot as plt

#Consider the following two functions representing the running times for some functions
def g(n):
    return n**2

def f(n):
    return n**3

#Creating list of values yielded based on the functions
x_vals = list(range(50))
f = [f(x) for x in x_vals]
g = [g(x) for x in x_vals]

#Plotting functions using matplotlib
plt.plot(x_vals, f, color='r', label='f(n)')
plt.plot(x_vals, g, color='b', label='g(n)')

plt.xlabel('Input size (n)')
plt.ylabel('Running Time T(n)')
plt.legend()

plt.show()
```

### Example:
As in the case of Big-O notation, we took two functions with running times n$^2$ and n$^3$ respectively. Now, as the size of input increase, it is clear that the output for n$^2$ will increase more slowly than the output of n$^2$. Since, we have g(n) as a lower bound over f(n), thus, we can say for the provided example:

> f(n) = $\Omega$(g(n)) $\forall$ n > 1

## $\Theta$ - Notation:
A function f(n) is said to belong to the set $\Theta$(g(n)) if there exist positive constants c$_1$ and c$_2$ such that:

> 0 $\leq$ c$_1$g(n) $\leq$ f(n) $\leq$ c$_2$g(n) ; $\forall$ n $\epsilon$ N, n $\geq$ n$_o$

We illustrate a simple graph for this expression using a simple example below:

```python
#Importing libraries
from matplotlib import pyplot as plt

#Consider the following two linear functions representing the running times for some functions
def f(n):
    return n + 50

def g(n):
    return n

#Creating list of values yielded based on the functions
x_vals = list(range(50))
lower_g = [g(x) for x in x_vals]
f = [f(x) for x in x_vals]
upper_g = [3*g(x) for x in x_vals]

#Plotting functions using matplotlib
plt.plot(x_vals, lower_g, color='r', label='c1g(n)')
plt.plot(x_vals, f, color='b', label='f(n)')
plt.plot(x_vals, upper_g, color='r', label='c2g(n)')

plt.xlabel('Input size (n)')
plt.ylabel('Running Time T(n)')
plt.legend()

plt.show()
```

### Example:
In the above provided plot, we can see that since the input size starts with 0 we have:

f(n) = 0 + 100 = 100 > c2g(n) = 2 * 0 = 0 > c1g(n) = 0

As we progress towards greater values, we reach a point i.e. approx n = 25 after which c2g(n) surpasses the f(n) function and forms upper bound on it whereas the c1g(n) function forms lower bound on it i.e.

c2g(n) = 3 * 25 = 125 > f(n) = 25 + 100 = 125 > c1g(n) = 25

Thus, we can say that the function f(n) is bounded above by (3 * n) and bounded below by (n), thus, if we ignore constants we get the function f(n) bounded both above and below by n. Thus, we can say for the given case:

>f(n) = $\Theta$(g(n)) $\forall$ n > 25

# Theorem of Asymptotic Notations:
For any two functions f(n) and g(n), we have:

> f(n) = $\Theta$(g(n)) iff f(n) = O(g(n)) = $\Omega$(g(n))

## Small o-Notation:
A function f(n) = o(g(n)) if and only if g(n) is loosely upper bounded on f(n) so that:

> 0 $\leq$ f(n) < cg(n) ; $\forall$ n $\epsilon$ N, n $\geq$ n$_o$

Thus, o(g(n)) implies that f(n) becomes insignificant relative to g(n) as n $\rightarrow$ $\infty$ i.e.

> lim$_{n \rightarrow \infty}$ f(n)/g(n) = 0

## Small $\omega$-Notation:
A function f(n) = $\omega$(g(n)) if and only if g(n) is loosely lower bounded on f(n) so that:

> 0 $\leq$ cg(n) < f(n) ; $\forall$ n $\epsilon$ N, n $\geq$ n$_o$

Thus, $\omega$(g(n)) implies that f(n) becomes insignificant large to g(n) as n $\rightarrow$ $\infty$ i.e.

> lim$_{n \rightarrow \infty}$ f(n)/g(n) = $\infty$

# Properties of Asymptotic Notations:
Asymptotic Notations all statisy some properties which are illustrated as follows:

### Transitivity:
This property is applicable on all asymptotic notaions. If

> f(n) = O(g(n)) and g(n) = O(h(n))

then
> f(n) = O(h(n))

### Relativity:
This property applies only to Big Notations i.e. O, $\Theta$ and $\Omega$

> f(n) = O(f(n)) 

### Symmetry:
This property applies only to $\Theta$ notation.

> f(n) = $\Theta$(g(n)) iff g(n) = $\Theta$(f(n))

### Transpose Symmetry:
This property applies to asymptotic notations with small forms only;

> f(n) = O(g(n)) iff g(n) = $\Omega$(f(n))

> f(n) = o(g(n)) iff g(n) = $\omega$(f(n)) 

# Some Basic Algorithm Operations:

## Monotonicity:
Monotonicity is the property of a function f to either perform directly proportional or inversely proportional relation with itself based on the value of the input.

### Monotonically Increasing Function:
A function f(n) is said to be monotonically increasing in nature if m $\leq$ n i.e. 

> f(m) $\leq$ f(n)

### Monotonically Decreasing Function:
A function f(n) is said to be monotonically decreasing in nature if m $\geq$ n i.e.

> f(m) $\geq$ f(n)

### Strictly Increasing Function:
A function f(n) is said to be strictly increasing in nature if m < n i.e.

> f(m) < f(n)

### Strictly Decreasing Function:
A function f(n) is said to be strictly decreasing in nature if m > n i.e.

> f(m) > f(n)

## Floor Function:
For any real number x, floor function is defined as the greatest integer less than or equal to x, denoted by $\lfloor$x$\rfloor$. For any provided float or integer we have:

> x - 1 < $\lfloor$x$\rfloor$ $\leq$ x

## Ceiling Function:
For any real number x, ceiling function is defined as the least integer greater than or equal to x, denoted by $\lceil$x$\rceil$. For any provided float or integer we have:

> x $\leq$ $\lceil$x$\rceil$ < x + 1

### Properties of Floor/Ceiling Function:
Following are some properties of floor and ceiling functions:

* $\lfloor$n/2$\rfloor$ + $\lceil$n/2$\rceil$ = n; $\forall$ n $\epsilon$ Z
* $\lceil$$\lceil$x/a$\rceil$/b$\rceil$ = $\lfloor$x/ab$\rfloor$; $\forall$ x $\epsilon$ R, a,b $\epsilon$ Z
* $\lfloor$$\lfloor$x/a$\rfloor$/b$\rfloor$ = $\lceil$x/ab$\rceil$; $\forall$ x $\epsilon$ R, a,b $\epsilon$ Z
* $\lceil$a/b$\rceil$ $\leq$ (a + (b - 1))/b; $\forall$ a,b $\epsilon$ Z
* $\lfloor$a/b$\rfloor$ $\geq$ (a - (b - 1))/b; $\forall$ a,b $\epsilon$ Z

Using the above properties, we find that:

* f(x) = $\lfloor$x$\rfloor$ is monotonically increasing 
* f(x) = $\lceil$x$\rceil$ is monotonically decreasing

## Modular Function:
The value of a mod n for any integer a and positive integer n is the remainder or residue of the quotient a/n i.e.

> f(n) = a mod n = a - n $\lfloor$a/n$\rfloor$ ; $\forall$ 0 $\leq$ f(n) < n

Thus, modular function can give the same result for two different values given that the numbers provided are constant multiples of each other.

### Example:
* Consider the following different values and their mods:

|Numbers|Mod 5|
|-|-|
|5,10,15,20,25|0|
|6,11,16,21,26|1|
|7,12,17,22,27|2|
|8,13,18,23,28|3|
|9,14,19,24,29|4|

* Similarly, we have for mods of 3:

|Numbers|Mod 3|
|-|-|
|3,6,9,12,15|0|
|4,7,10,13,16|1|
|5,8,11,14,17|2|

From the above two examples, we can observe that:

* a mod n has atmost n answers after which the cycle is repeated (0 to n-1)
* a $\equiv$ b (mod n) iff n is a divisor of (b - a). This is termed as __Equivalency Property of Modulo__.

## Polynomial Function:
Suppose we are provided a polynomial in n of some degree d, where d > 0, d $\epsilon$ Z, then if p(n) be the function of the polynomial we have:

> p(n) = $\Sigma_{i = 0}^{d} a_{i} n^i$ ; $\forall$ a$_i$ are coefficients of p(n) and a$_d$ = 0

### Asymptotic Positive Polynomial Function:
A polynomial is said to be asymptotically positive in nature if and only if a$_d$ > 0. Thus, we have for the asymptotic complexity of such polynomial function:

> p(n) = $\Theta$(n$^d$)

For a given function f(n), it is said to be polynomially bounded if we have:

> f(n) = O(n$^k$) ; $\forall$ k = constant

## Exponential Function:
A function is said to be exponential for any a > 0:

> f(n) = a$^n$ ; $\forall$ n $\epsilon$ Z

### Properties of Exponential Function:
For real integers m,n, we have the following properties for exponential function:

* a$^o$ = 1
* a$^1$ = a
* a$^{-1}$ = 1/a
* (a$^m$)$^n$ = a$^{mn}$ = (a$^n$)$^m$
* a$^m$ * a$^n$ = a$^{m + n}$

It is clear that we have for any particular a:

a$^o$ = 1 < a$^1$ = a < a$^2$ < a$^3$ < ... < a$^{m-1}$ < a$^m$

Thus, exponential function is monotonically increasing in n for all n where a $\geq$ 1

## Logarithmic Function:
Logarithmic functions are defined using the following formula:

> f(n) = log$_{base}$ n ; $\forall$ n $\epsilon$ R$^+$ 

Based on the different bases, we have:

* log$_2$ n (Base = 2) => Binary Logarithm
* log$_e$ n (Base = e) = ln n => Natural Logarithm
* log$_{10}$ n (Base = 10) => Common Logarithm
* log$^k$ n (Base = 2) => Binary Exponentation Logarithm
* lg(lg n) (Base = 2) => Binary Composition Logarithm

### Properties of Logarithmic Function:
Suppose we have a > 0, b > 0 and c > 0, then for a particular positive n, we have:

* a = b$^{log_{b}a}$
* log$_c$ ab = log$_c$ a + log$_c$b
* log$_b$a$^n$ = nlog$_b$a
* log$_b$a = log$_c$a / log$_c$b
* log$_b$(1/a) = log$_b$(a)$^{-1}$ = -log$_b$a (Using log$_b$a$^n$ = nlog$_b$a)
* log$_b$a = 1/log$_a$b
* a$^{log_bc}$ = c$^{log_ba}$

## Polylogarithmic Functions:
A function f(n) is said to be polylogarithmic in nature if and only if it is bounded as:

> f(n) = O(log$^k$ n) ; $\forall$ k = constant

## Factorial Function:
Factorial is a function f(n) denoted by the symbol '!' and is defined as :

> f(n) = n! { 1 if n = 0 and n(n-1)! if n > 0

Thus, we can say that:

> n! = 1 * 2 * 3 * ... * (n - 1) * n

### Stirling's Approximation for Factorial Function:
Stirling proposed a formula based on the property that there is a weak upper bound on the factorial function i.e. n! $\leq$ n$^n$. The formula is known as "Stirling's Approximation" and is given as:

> n! = $\sqrt{2 \pi n}$ (n/e)$^n$ (1 + $\Theta$(1/n)) ; where n! = o(n$^n$) = $\omega$(2$^n$)

Using this approximation, we can deduce the formula for logarithm of a factorial as:

> log(n!) = $\Theta$(nlogn)

Similarly, we can say that:

> n! = $\sqrt{2 \pi n}$ (n/e)$^n$ e$^{\alpha_n}$ ; $\forall$ 1/(12n + 1) < $\alpha_n$ < (1/12n) , n $\geq$ 1

## Functional Iteration:
An iterative function f(n) iterated i times to an initial value n, is denoted by f^${(i)}$(n) and is defined as:

> f^{(i)}(n) = { n if i = 0 and f(f$^{i-1}$(n)) if i > 0

### Iterated Logarithm Function:
An iterated logarithm function us denoted by log$^*$n and is defined as:

> log^$*$n = min {i $\geq$ 0: log$^{(1)}$n $\leq$ 1}

#### Example:
The iterated logarithm function has very very slow growth rate overall i.e. we have:

* log$^*$2 = 1
* log$^*$4 = 2
* log$^*$16 = 3
* log$^*$65536 = 4
* log$^*$(2$^{65536}$) = 5 > 10$^{80}$ (No. of atoms in universe i.e. we cannot get such high outputs)

This shows that the denominator > numbertor i.e. n! > n$^3$

#### Which is asymptotically larger lg(lg$^*$n) or lg$^*$(lg n)?
